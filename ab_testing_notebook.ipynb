# 📊 A/B Testing Analysis — Website Background Experiment


## 🎯 Objective
Evaluate whether changing website background color (White → Black) improves conversion rate without harming engagement.


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.proportion import proportions_ztest, confint_proportions_2indep
```

## 📥 Load Data

```

df = pd.read_csv('ab_testing.csv')
df.head()
```

## 🔄 Data Preparation

```

df['converted'] = (df['Conversion'] == 'Yes').astype(int)
df.head()
```

## 📊 Summary Metrics

```

summary = df.groupby('Group').agg(
    users=('User ID', 'count'),
    conversions=('converted', 'sum'),
    conversion_rate=('converted', 'mean'),
    avg_time_spent=('Time Spent', 'mean'),
    avg_page_views=('Page Views', 'mean')
)

summary
```

## 📈 Conversion Rate Visualization

```

summary['conversion_rate'].plot(kind='bar')
plt.title('Conversion Rate by Group')
plt.ylabel('Conversion Rate')
plt.show()
```

## 🔬 Statistical Test (Z-test)

```

conv = summary['conversions'].values
nobs = summary['users'].values

z_stat, p_value = proportions_ztest(conv, nobs)

print("Z-stat:", z_stat)
print("P-value:", p_value)
```

## 📏 Confidence Interval

```

low, high = confint_proportions_2indep(
    count1=conv[1], nobs1=nobs[1],
    count2=conv[0], nobs2=nobs[0],
    method='wald'
)

print("CI:", (low, high))
```


## ⚖️ Decision

- If p-value < 0.05 → significant  
- If CI does not include 0 → reliable uplift  

👉 Based on results:
- Significant improvement in conversion  
- No negative impact on guardrail metrics  

✅ **Decision: Roll out Treatment (Black background)**
