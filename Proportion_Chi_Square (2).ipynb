```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import scipy.stats as stats

# Population 1 = Jakarta
# Population 2 = Bandung
# Population 3 = Yogyakarta

# Voucher A = minimum transaction 10k will get discount 1.0k
# Voucher B = minimum transaction 12k will get discount 1.2k

data = {
    'voucher':['A','B'],
    'population_1_redeem':[1161,1421],
    'population_1_not_redeem':[198788,198541],
    'population_2_redeem':[150,151],
    'population_2_not_redeem':[269604,269613],
    'population_3_redeem':[334,512],
    'population_3_not_redeem':[27333,25961]
}

df = pd.DataFrame(data)

populations = ['population_1','population_2','population_3']
results = []
alpha = 0.05

for pop in populations:
    redeem_col = pop + '_redeem'
    not_redeem_col = pop + '_not_redeem'

    redeem_A = df[df['voucher'] == 'A'][redeem_col].values[0]
    redeem_B = df[df['voucher'] == 'B'][redeem_col].values[0]
    not_redeem_A = df[df['voucher'] == 'A'][not_redeem_col].values[0]
    not_redeem_B = df[df['voucher'] == 'B'][not_redeem_col].values[0]

    observed = np.array([
        [redeem_A, not_redeem_A],
        [redeem_B, not_redeem_B]
    ])

    chi2, p, dof, _ = chi2_contingency(observed, correction=True)

    result = {
        'Population': pop,
        'Total Redeem A': redeem_A,
        'Total Redeem B': redeem_B,
        'P-value': p,
        'Reject Null Hypothesis': p < alpha
    }

    results.append(result)

result_df = pd.DataFrame(results)
result_df

# H0 = no significant difference
# H1 = significant difference
```
