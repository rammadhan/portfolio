# A/B Testing Analysis — Website Background Experiment


## Objective
Evaluate whether changing website background color black to white improves conversion rate without harming engagement.

Group A = White color (Control)

Group B = Black color (treatment)

data_source: https://www.kaggle.com/datasets/adarsh0806/ab-testing-practice/data

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.proportion import proportions_ztest, confint_proportions_2indep
```

## Load Data

```python

df = pd.read_csv('ab_testing.csv')
df.head()
```

 Data Preparation

```python

df['converted'] = (df['Conversion'] == 'Yes').astype(int)
df.head()
```

##  Summary Metrics

```python

summary = df.groupby('Group').agg(
    users=('User ID', 'count'),
    conversions=('converted', 'sum'),
    conversion_rate=('converted', 'mean'),
    avg_time_spent=('Time Spent', 'mean'),
    avg_page_views=('Page Views', 'mean')
)

summary
```

Statistical Test (Z-test)

```python

conv = summary['conversions'].values
nobs = summary['users'].values

z_stat, p_value = proportions_ztest(conv, nobs)

print("Z-stat:", z_stat)
print("P-value:", p_value)
```

## 📏 Confidence Interval

```python

low, high = confint_proportions_2indep(
    count1=conv[1], nobs1=nobs[1],
    count2=conv[0], nobs2=nobs[0],
    method='wald'
)

print("CI:", (low, high))
```


##  Sumary

The treatment variant significantly improves conversion rate by +8.67% (95% CI: +7.0% to +10.3%, p < 0.001), with no negative impact on engagement metrics, supporting a full rollout.
